```python
import numpy as np
import urllib
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
response = urllib.request.urlopen(url)
```

```python
data = np.genfromtxt(response, delimiter=",", usecols=(0, 1, 2, 3))
data[:10]
```

```python
print(data.ndim)
print(data.shape)
```

```python
print(f"sepal length mean: {np.mean(data[:, 0])}")
print(f"sepal length median: {np.median(data[:, 0])}")
print(f"sepal length max: {np.max(data[:, 0])}")
print(f"sepal length min: {np.min(data[:, 0])}")
print(f"sepal length std: {np.std(data[:, 0])}")
```

```python
print(f"sepal width mean: {np.mean(data[:, 1])}")
print(f"sepal width median: {np.median(data[:, 1])}")
print(f"sepal width max: {np.max(data[:, 1])}")
print(f"sepal width min: {np.min(data[:, 1])}")
print(f"sepal width std: {np.std(data[:, 1])}")
```

```python
print(f"sepal width mean: {np.mean(data[:, 1]):.2f}")
print(f"sepal width median: {np.median(data[:, 1]):.2f}")
print(f"sepal width max: {np.max(data[:, 1]):.2f}")
print(f"sepal width min: {np.min(data[:, 1]):.2f}")
print(f"sepal width std: {np.std(data[:, 1]):.2f}")
```

```python
print("petal length statistics:")
print(f"mean rounded: {np.mean(data[:, 2]):.2f}")
print(f"median: {np.median(data[:, 2]):.2f}")
print(f"max: {np.max(data[:, 2]):.2f}")
print(f"min: {np.min(data[:, 2]):.2f}")
print(f"std: {np.std(data[:, 2]):.2f}")
```

```python
PETAL_LENGTH = 2 # position

max_petal_length_index = np.argmax(data[:, PETAL_LENGTH])

max_petal_length = data[max_petal_length_index, PETAL_LENGTH] # data [0, 0]

print("Índice de la longitud del pétalo más larga:", max_petal_length_index)
print("Longitud del pétalo más larga:", max_petal_length)
```

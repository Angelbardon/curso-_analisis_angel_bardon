### Correlación

```python
df_corr = df.corr(numeric_only=True)
sns.heatmap(df_corr,annot=True,cmap='viridis')
```

```python
import numpy as np
df_corr = df.corr(numeric_only=True)
filter = np.triu(df_corr)
sns.heatmap(df_corr,annot=True, cmap='viridis', mask= filter)
```

```python
plt.figure(figsize=(20,15), dpi=300)
df = pd.get_dummies(df)
df_corr = df.corr()
filter = np.triu(df_corr)
sns.heatmap(df_corr,annot=True, cmap='viridis', mask= filter, fmt='.2f'); #ftm='.2f' saca solo 2 decimales
```

```python
# CORRELACION PARA LA COLUMNA PRICE ***IMPORTANTE***
df_corr = df.corr() # matriz de correlación
corr_price = df_corr['price']# filtrar correlacion precio
corr_price = corr_price.sort_values(ascending=False)[1:] # ordenar descendente
plt.figure(figsize=(15,5))
sns.barplot(x=corr_price.index, y=corr_price.values)
plt.xticks(rotation=45);
```

```python
### CÓDIGO PARA REUTILIZAR ###

df_corr = df.corr() # matriz de correlación
corr_price = df_corr['price']# filtrar correlación precio
filter = (corr_price < -0.1) | (corr_price > 0.1) # quitar aquellas de baja correlación
corr_price = corr_price[filter]
corr_price = corr_price.sort_values(ascending=False)[1:] # ordenar descendente

plt.figure(figsize=(15,5))
ax = sns.barplot(x=corr_price.index, y=corr_price.values)
plt.xticks(rotation=45)

labels = [f'{value: .2f}' for value in corr_price.values] # redondear a dos decimales
ax.bar_label(ax.containers[0], labels=labels);
```
